```python
import pandas as pd
# from datetime import datetime as dt
import matplotlib.pyplot as plt
# import seaborn as sns 
# sns.set(style="darkgrid")
```

```python
def plot_figs(name_to_log_mapping):
    plt.figure(figsize=(20,10))
    for name, log in name_to_log_mapping.items():
        plt.plot(log['sys_time'], log['logical_time'], label=name+' logical', linewidth=2)
        plt.plot(log['sys_time'], log['queue_len'], label=name+' queue', linewidth=4)
    plt.legend(loc="upper left")
    plt.show()
```

```python
logs = {
    'm0': pd.read_csv('logs/122machine0.csv', skiprows=1),
    'm1': pd.read_csv('logs/122machine1.csv', skiprows=1),
    'm2': pd.read_csv('logs/122machine2.csv', skiprows=1)
}
plot_figs(logs)
```

```python
import pandas as pd
# from datetime import datetime as dt
import matplotlib.pyplot as plt
# import seaborn as sns 
# sns.set(style="darkgrid")
```

```python
def plot_figs(name_to_log_mapping, save_name=None):
    plt.figure(figsize=(20,10))
    for name, log in name_to_log_mapping.items():
        plt.plot(log['sys_time'], log['logical_time'], label=name+' logical', linewidth=2)
        plt.plot(log['sys_time'], log['queue_len'], label=name+' queue', linewidth=4, marker=".")
    plt.legend(loc="upper left")
#    plt.show()
    if save_name is not None:
        plt.savefig('figs/'+save_name+'.png')
    
```

```python
runsig = '115'
logs = {
    'm0': pd.read_csv(f'logs/{runsig}machine0.csv', skiprows=1),
    'm1': pd.read_csv(f'logs/{runsig}machine1.csv', skiprows=1),
    'm2': pd.read_csv(f'logs/{runsig}machine2.csv', skiprows=1)
}
plot_figs(logs, save_name=runsig)
```
